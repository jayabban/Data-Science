Splitting data

```python
X = df1.iloc[:, :-1].values
y = df1.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X[:, 0:] = sc.fit_transform(X[:, 0:])
X = np.array(sc.transform(X))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lg_classifier = LogisticRegression()
lg_classifier.fit(X_train, y_train)
```

```python
print("Accuracy: ", lg_classifier.score(X_test, y_test) * 100)
```

```python
lg_acc = (lg_classifier.score(X_test, y_test) * 100)
```

```python
lg_y_pred = lg_classifier.predict(X_test)
print(lg_y_pred)
```

```python
print(np.concatenate((lg_y_pred.reshape(len(lg_y_pred),1), y_test.reshape(len(y_test),1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, lg_y_pred)
print(cm)
```

```python
accuracy_score(lg_y_pred, y_test)
```

```python
lg_result = pd.DataFrame(['Logistic Regression', lg_acc]).transpose()
```

```python
lg_result
```

```python
label = ['M','B']
ax = sns.heatmap(cm, annot= True, fmt='d', cmap='Greens', xticklabels=label, yticklabels=label)
ax.set_title("Confusion Metrics")
ax.set_xlabel('Predict')
ax.set_ylabel('Actual')
```
