# Trainning the model

## Feature Selection

```python
df.head()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

## Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc =StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(X_train, y_train)
```

```python
kn_pred = kn.predict(X_test)
print('Accuracy: ', kn.score(X_test, y_test) * 100)
```

```python
kn_pred_acc = print('Accuracy: ', kn.score(X_test, y_test) * 100)
kn_pred_acc
```

```python
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, kn_pred)
print(cm)
```

```python
# print(np.concatenate((y_test.reshape(len(y_test),1), kn_pred.reshape(len(kn_pred),1)), 1))
```

```python
ax = sns.heatmap(cm, annot=True, fmt='d',cmap='Greens')
plt.title("Confusion Matrix")
ax.set_xlabel("Predict")
ax.set_ylabel("Actual")
```
